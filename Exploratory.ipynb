```python
country_acronyms = {
    'Poland': 'PL',
    'Romania': 'RO',
    'Spain': 'ES',
    'Greece': 'GR',
    'Germany': 'DE',
    'Belgium': 'BE',
    'Austria': 'AT',
    'Bulgaria': 'BG',
    'Latvia': 'LV',
    'Sweden': 'SE',
    'France': 'FR',
    'United Kingdom': 'GB',
    'Hungary': 'HU',
    'Netherlands': 'NL',
    'Denmark': 'DK',
    'Lithuania': 'LT',
    'Czech Republic': 'CZ',
    'Norway': 'NO',
    'Italy': 'IT',
    'Portugal': 'PT',
    'Malta': 'MT',
    'Cyprus': 'CY',
    'Slovakia': 'SK',
    'Iceland': 'IS',
    'Croatia': 'HR'
}

articles_set = set()
def article_indeces(articles):
    
    indeces = []
    
    for article in articles.split('|'): 
        parts = article.split(' ')
        if len(parts) > 1:
            try:
                indeces.append(int(parts[1]))
                articles_set.add(int(parts[1]))
            except ValueError:
                continue
            
    
    return "|" + "|".join(map(str, indeces)) + "|"

gdpr_violations['country_acronym'] = gdpr_violations['name'].map(country_acronyms)
gdpr_violations['article_ids'] = gdpr_violations['article_violated'].apply(lambda x: article_indeces(x) if isinstance(x, str) else [])
gdpr_violations['date_iso'] = gdpr_violations['date'].apply(lambda x: "-".join(x.split("/")[::-1]))
gdpr_violations.drop_duplicates(subset=['id'], inplace=True, keep='first')
gdpr_violations.to_csv("./website/assets/data/gdpr_violations.csv", index=False)
```
